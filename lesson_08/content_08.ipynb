2. `**kwargs` –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –¥–ª—è –ø–µ—Ä–µ–¥–∞—á—ñ –¥–æ–≤—ñ–ª—å–Ω–æ—ó –∫—ñ–ª—å–∫–æ—Å—Ç—ñ —ñ–º–µ–Ω–æ–≤–∞–Ω–∏—Ö –∞—Ä–≥—É–º–µ–Ω—Ç—ñ–≤ (–∫–ª—é—á-–∑–Ω–∞—á–µ–Ω–Ω—è):
    
    ```python
    def print_kwargs(**kwargs):
        for key, value in kwargs.items():
            print(f"{key}: {value}")
    
    # –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–ª–∏–∫—É —Ñ—É–Ω–∫—Ü—ñ—ó
    print_kwargs(name="John", age=25, city="New York")
    ```
    
3. –ö–æ–º–±—ñ–Ω–∞—Ü—ñ—è `*args` —ñ `**kwargs` –≤ —Ç—ñ–ª—ñ —Ñ—É–Ω–∫—Ü—ñ—ó –¥–æ–∑–≤–æ–ª—è—î –ø—Ä–∏–π–º–∞—Ç–∏ –±—É–¥—å-—è–∫—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å –ø–æ–∑–∏—Ü—ñ–π–Ω–∏—Ö —Ç–∞ —ñ–º–µ–Ω–æ–≤–∞–Ω–∏—Ö –∞—Ä–≥—É–º–µ–Ω—Ç—ñ–≤:
    
    ```python
    def print_args_and_kwargs(*args, **kwargs):
        for arg in args:
            print(arg)
        
        for key, value in kwargs.items():
            print(f"{key}: {value}")
    
    # –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–ª–∏–∫—É —Ñ—É–Ω–∫—Ü—ñ—ó
    print_args_and_kwargs(1, "hello", 3.14, name="John", age=25)
    ```

## –õ—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—ó

–õ—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—ó (—Ç–∞–∫–æ–∂ –≤—ñ–¥–æ–º—ñ —è–∫ –∞–Ω–æ–Ω—ñ–º–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—ó) –≤ Python - —Ü–µ –∫–æ—Ä–æ—Ç–∫—ñ —Ñ—É–Ω–∫—Ü—ñ—ó, —è–∫—ñ –º–æ–∂–Ω–∞ –≤–∏–∑–Ω–∞—á–∏—Ç–∏ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –∫–ª—é—á–æ–≤–æ–≥–æ —Å–ª–æ–≤–∞ `lambda`. –í–æ–Ω–∏ —á–∞—Å—Ç–æ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—Ç—å—Å—è –¥–ª—è –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è –º–∞–ª–∏—Ö —Ñ—É–Ω–∫—Ü—ñ–π –Ω–∞ –ª—å–æ—Ç—É, –¥–µ –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è –ø–æ–≤–Ω–æ—Ü—ñ–Ω–Ω–æ—ó —Ñ—É–Ω–∫—Ü—ñ—ó –Ω–µ —î –Ω–µ–æ–±—Ö—ñ–¥–Ω–∏–º.

–°–∏–Ω—Ç–∞–∫—Å–∏—Å –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—ó –≤–∏–≥–ª—è–¥–∞—î –Ω–∞—Å—Ç—É–ø–Ω–∏–º —á–∏–Ω–æ–º:

```python
lambda arguments: expression
```

**–ü—Ä–∏–∫–ª–∞–¥ –∑ –æ–¥–Ω–∏–º –∞—Ä–≥—É–º–µ–Ω—Ç–æ–º:**

```python
square = lambda x: x**2
print(square(5))  # –í–∏–≤–µ–¥–µ: 25
```
